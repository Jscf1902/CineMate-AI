```python
import pandas as pd
import requests
import uuid
import random
```

```python
def cargar_preguntas(ruta_csv):
    df = pd.read_csv(ruta_csv)
    return df
```

```python
def generar_telefono():
    numero = "5731" + str(random.randint(10000000, 99999999))
    return numero
```

```python
def generar_session_id():
    return str(uuid.uuid4())
```

```python
def enviar_pregunta(url, pregunta, telefono, session_id):    
    payload = {
        "session_id": session_id,
        "user_id": telefono,
        "message": pregunta
    }
    
    try:
        response = requests.post(url, json=payload)
        response.raise_for_status()
        return response.json().get("response", "")
    
    except Exception as e:
        return "ERROR"
```

```python
def procesar_preguntas(df, url):
    import pandas as pd
    
    resultados = []
    
    for i, row in df.iterrows():
        pregunta = row["Pregunta"]
        tipo = row["Tipo"]
        
        telefono = generar_telefono()
        session_id = generar_session_id()
        
        print("\n==============================")
        print(f"Iteración: {i+1}")
        print(f"Teléfono: {telefono}")
        print(f"Pregunta: {pregunta}")
        
        respuesta = enviar_pregunta(
            url,
            pregunta,
            telefono,
            session_id
        )
        
        print(f"Respuesta: {respuesta}")
        print("==============================\n")
        
        resultados.append({
            "pregunta": pregunta,
            "tipo": tipo,
            "telefono": telefono,
            "respuesta": respuesta
        })
    
    return pd.DataFrame(resultados)
```

```python
def guardar_resultados(df, ruta_salida):
    df.to_csv(ruta_salida, index=False)
```

```python
# 1. Ruta del CSV
ruta = r"C:\Users\Juan1902\Downloads\Test\test_questions.csv"

# 2. URL del endpoint
url = "http://127.0.0.1:8000/chat"

# 3. Cargar preguntas
df_preguntas = cargar_preguntas(ruta)

# 4. Procesar
df_resultados = procesar_preguntas(df_preguntas, url)

# 5. Guardar resultados
guardar_resultados(df_resultados, "resultados_pruebas.csv")

# 6. Ver resultados
df_resultados.head()# 1. Ruta del CSV
ruta = r"C:\Users\Juan1902\Downloads\Test\test_questions.csv"

# 2. URL del endpoint
url = "http://127.0.0.1:8000/chat"

# 3. Cargar preguntas
df_preguntas = cargar_preguntas(ruta)

# 4. Procesar
df_resultados = procesar_preguntas(df_preguntas, url)

# 5. Guardar resultados
guardar_resultados(df_resultados, "resultados_pruebas.csv")

# 6. Ver resultados
df_resultados.head()
```
